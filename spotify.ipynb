```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("spotify-2023.csv",encoding = 'latin-1')
```

```python
data.nunique()
```

```python
data.head(30)
```

```python
data.isna().sum()
```

```python
data = data.drop(574)
```

```python
data['streams'] = data['streams'].astype(np.int64)
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='mode')
plt.title('Number of songs by mode')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=data, x='key')
plt.title('Number of songs by key')
plt.xticks(rotation=90)
plt.show()
```

```python
# Correlation Analysis
correlation_matrix = data[['in_spotify_charts', 'in_apple_charts']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Danceability
plt.figure(figsize=(12, 6))
sns.boxplot(data=data, x='key', y='danceability_%')
plt.title('Danceability by key')
plt.xticks(rotation=90)
plt.show()
```

```python
# Release Date Analysis
plt.figure(figsize=(8, 6))
sns.lineplot(data=data, x='released_year', y='streams')
plt.title('Release years vs streams')
plt.xticks(rotation=90)
plt.gca().set_yticklabels([])
plt.show()
```

```python
#Top 300 streamed songs in spotify
data300 = data.sort_values(by="streams",ascending=False,inplace=True)
```

```python
#what is the most used key in the first 300 songs
data300 = data[:300]
plt.figure(figsize=(8, 6))
sns.countplot(data=data300, x='key')
plt.title('Number of songs by key')
plt.xticks(rotation=90)
plt.show()
```

```python
#what is the most used key in the first 100 songs
data100 = data[:100]
plt.figure(figsize=(8, 6))
sns.countplot(data=data100, x='key')
plt.title('Number of songs by key')
plt.xticks(rotation=90)
plt.show()
```
